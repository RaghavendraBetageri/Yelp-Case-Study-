Find the top business categories based on the total number of reviews. Output the category along with the total number of reviews. Order by total reviews in descending order.

```python
import pandas as pd
```

```python
df = pd.read_excel("Yelpbusiness.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[["categories", "review_count"]]
```

```python
dic = {}
for i in df.iterrows():
    for j in (i[1][0]).split(";"):
        if j in dic:
            dic[j] += i[1][1]
        else:
            dic[j] = i[1][1]
```

```python
sorted(list(dic.items()), key=lambda x: x[1], reverse=True)
```
